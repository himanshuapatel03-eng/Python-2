# Pagination

```python
import requests
url = 'https://jsonplaceholder.typicode.com/posts'
params = {
    '_page':1,
    '_limit':10
}
response = requests.get(url,params=params)
data = response.json()
for post in data:
    print(post['id'],post['title'])
```

```python
import requests
url = 'https://jsonplaceholder.typicode.com/posts'
page = 1
while page<=3:
    params = {
        '_page':page,
        '_limit':10
    }
    response = requests.get(url,params=params)
    data = response.json()
    for post in data:
        print(post['id'],post['title'])
    page+=1
```

# Token ( Authorization )

```python
import requests
url = 'https://jsonplaceholder.typicode.com/posts'
token = 'my_secret_token'
headers = {'Authorization':f'Bearer {token}'}
response = requests.get(url,headers=headers)
print(response.status_code)
print(response.json())
```
