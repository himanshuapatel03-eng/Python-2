# Web Scraping

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://subslikescript.com/movie/Titanic-120338'
result = requests.get(url)
content = result.text

soup = BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
soup.find('h1').text
```

```python
len(soup.find_all('a',href=True))
```

```python
movie = soup.find('article',class_='main-article')
movie.find('h1')
```

```python
script = movie.find('div',class_='full-script').text.strip()
print(script)
```

```python
title = soup.find('h1').text.strip().split()[0]
title
```

```python
with open(f'{title}.txt','w',encoding='utf-8') as file:
    file.write(script)
```

### Task

```python
url = 'https://subslikescript.com/movies'
result = requests.get(url)
content = result.text

soup = BeautifulSoup(content,'html.parser')
```

```python
movie = soup.find('ul',class_='scripts-list')

links = []
for i in movie.find_all('a',href=True):
    links.append('https://subslikescript.com'+i['href'])

for l,i in zip(links,movie.find_all('a',href=True)):
    result = requests.get(l)
    content = result.text

    soup = BeautifulSoup(content,'html.parser')

    box = soup.find('article',class_='main-article')
    
    title = i['href'][7:]

    script = box.find('div',class_='full-script').text

    with open(f'{title}.txt','w',encoding='utf-8') as file:
        file.write(script)
```
