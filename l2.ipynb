### Task 2

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://politifact.com/factchecks'
result = requests.get(url)
content = result.text

soup = BeautifulSoup(content,'html.parser')

author = soup.find_all('a',class_='m-statement__name')
quote = soup.find_all('div',class_='m-statement__quote')
date = soup.find_all('footer')
authors = []
quotes = []
links = []
dates = []


for i,j,k in zip(author,quote,date):
    authors.append(i.text.strip())
    quotes.append(j.text.strip())
    links.append('https://politifact.com'+j.find('a')['href'])
    dates.append(k.text.split('•')[-1].strip())

df = pd.DataFrame({'Author':authors,'Quote':quotes,'Link':links,'Date':dates})
df
```

### Access multiple pages

```python
authors = []
quotes = []
links = []
dates = []
for i in range(0,5):
    url = f'https://politifact.com/factchecks/?page={i}'

    result = requests.get(url)
    content = result.text
    
    soup = BeautifulSoup(content,'html.parser')
    
    author = soup.find_all('a',class_='m-statement__name')
    quote = soup.find_all('div',class_='m-statement__quote')
    date = soup.find_all('footer')
    
    for i,j,k in zip(author,quote,date):
        authors.append(i.text.strip())
        quotes.append(j.text.strip())
        links.append('https://politifact.com'+j.find('a')['href'])
        dates.append(k.text.split('•')[-1].strip())
    
df = pd.DataFrame({'Author':authors,'Quote':quotes,'Link':links,'Date':dates})
df
```

### Task 3

```python
soup = BeautifulSoup(open('top250.html','r',encoding='utf-8'))

title = soup.find_all('td',class_='titleColumn')
rating = soup.find_all('td',class_='ratingColumn imdbRating')
titles = []
years = []
ratings = []

for i,j in zip(title,rating):
    titles.append(i.find('a',href=True).text)
    years.append(i.find('span').text[1:-1])
    ratings.append(j.text.strip())

df = pd.DataFrame({'Title':titles,'Year':years,'Rating':ratings})
df
```

### Task 4

```python
soup = BeautifulSoup(open('flipkart.html','r',encoding='utf-8'))

name = soup.find_all('div',class_='KzDlHZ')
ram = soup.find_all('div',class_='_6NESgJ')
rating = soup.find_all('div',class_='XQDdHH')
price = soup.find_all('div',class_='Nx9bqj _4b5DiR')
names = []
colors = []
RAM = []
ROM = []
ratings = []
prices = []

for i,j,k,l in zip(name,ram,rating,price):
    names.append(i.text.split('(')[0])
    colors.append(i.text.split('(')[1].split(',')[0])
    ROM.append(i.text.split('(')[1].split(',')[1][1:-1])
    RAM.append(j.find('li').text.split('|')[0][0:-5])
    ratings.append(k.text)
    prices.append(l.text)

df = pd.DataFrame({'Name':names,'Color':colors,'RAM':RAM,'ROM':ROM,'Rating':ratings,'Price':prices})
df
```
